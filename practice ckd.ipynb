```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("kidney_disease.csv")
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df["rc"].unique()
```

```python
df["rc"] = df["rc"].replace("\t?", '0')
```

```python
df["rc"].unique()
```

```python
df["rc"] = df["rc"].astype(float)
```

```python
df["wc"].unique()
```

```python
df["wc"] = df["wc"].replace(("\t8400", "\t6200", "\t?"), (8400, 6200, 0))
```

```python
df["wc"] = df["wc"].astype(float)
```

```python
df["pcv"].unique()
```

```python
df["pcv"] = df["pcv"].replace(("\t?", "\t43"), (0, 43))
df["pcv"] = df["pcv"].astype(float)
```

```python
df.info()
```

```python
df.drop(["id"], axis = 1, inplace = True)
```

```python
df.describe()
```

```python
obj = (df.select_dtypes(include = object)).columns
numeric = (df.select_dtypes(include = np.number)).columns
print(numeric, obj)
```

```python
for i in obj:
    print(i)
    print(df[i].unique())
```

```python
df["dm"] = df["dm"].replace({"\tno":"no", "\tyes":"yes", " yes":"yes"})
df["cad"] = df["cad"].replace("\tno", "no")
df["classification"] = df["classification"].replace("ckd\t", "ckd")
```

```python
for i in obj:
    print(i)
    print(df[i].unique())
```

## for missing values

```python
## Dealing with missing values
df.isnull().sum()
```

```python
df["rc"].fillna(df["rc"].mean(), inplace=True)
```

```python
#filling numeric missing values with mean values in columns:
for i in numeric:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
#filling categorical values with mode values in columns
for i in obj:
    df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
df.head(10)
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in obj:
    le = LabelEncoder()
    df[i] = le.fit_transform(df[i])
    
    
```

```python
y = df["classification"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
x_train.shape, x_test.shape, y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
ran_fore = RandomForestClassifier(n_estimators=100)
ran_fore.fit(x_train, y_train)
rf_pred = ran_fore.predict(x_test)
rf_score = round(ran_fore.score(x_train, y_train)*100, 2)
rf_test_score = round(ran_fore.score(x_test, y_test)*100, 2)
print("Random forest train score = \n", rf_score)
print("Random forest test score = \n", rf_test_score)
print("accuracy  = \n", accuracy_score(y_test,rf_pred))
print("confusion Matrix : \n", confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```
